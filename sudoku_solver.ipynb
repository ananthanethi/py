```python
def col_check(col,key):
       flag=0
       for i in range(9):
            if df1[i][col]==key:
                    flag=1
       if flag==1:
            return False
       else:
           return True
    
```

```python
def row_check(row,key):
       flag=0
       for i in range(9):
            if df1[row][i]==key:
                    flag=1
       if flag==1:
            return False
       else:
           return True
```

```python
import math
def box_check(row,col,key):
        start_row=(math.floor(row/3)*3)
        end_row=start_row+2
        start_col=(math.floor(col/3)*3)
        end_col=(math.floor(col/3)*3)+2
        flag=0
        for i in range(start_row,end_row):
             for j in range(start_col,end_col):
                if df1[i][j]==key:
                    flag=1
        if flag==1:
            return False
        else:
            return True
```

```python
for i in range(9):
      for j in range(9):
            if (df1[i][j]==-1):
                       for k in range(9):
                            count=0
                            flag=0
                            if row_check(i,k+1):
                                if col_check(j,k+1):
                                    if box_check(i,j,k+1):
                                        count+=1
                                        if count>1:
                                             flag=1
                                         
                            if flag==0:                
                                df1[i][j]=k+1
                                        
                                                 
            
        
```
